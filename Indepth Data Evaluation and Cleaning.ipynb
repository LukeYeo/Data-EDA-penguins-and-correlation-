# 4. Correlations 

**Covariance**

It represents how two variables/columns/attributes are related. Higher values indicates higher dependency/relation between them. Values ranges from -ve infinity to +ve infinity. +ve values indicates direct relation (increase in x value will have increase in y value) and -ve value shows inverse relation (increase in x value will have decrease in y value).

```python
penguin_data.groupby('Species').cov()
```

<u>**Correlation**</u>

It represents the strength of the relation and it is standardized, values ranges between -1 to +1. Variables are strongly correlated when the value is close to +ve 1 (positive correlation) or -ve 1 (negative correlation). 0 indicates no relation among them.

**Pearason Correlation Coefficient (Pearson r):**

It measures the linear relation among the variables. Given by ratio of the covariance of x and y to the product of their standard deviations. https://en.wikipedia.org/wiki/Correlation

```python
penguin_data.groupby('Species').corr().style.background_gradient(cmap='Reds')
```

**Spearman Correlation Coefficient (Spearman rho):**

It measures the monotonic relation (non-linear relation). 
Same as pearson but uses ranked values instead of original values.
https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient

```python
penguin_data.groupby('Species').corr(method = 'spearman').style.background_gradient(cmap='Reds')
```

**Kendall Correlation Coefficient (Kendall tau):**

It measures the monotonic relation (non-linear relation). Similar to spearman based on ranked values and uses pair of observations to determine its relation. It is more robust and benifitical to use when there is a outliers in the data.
https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient

```python
penguin_data.groupby('Species').corr(method = 'kendall').style.background_gradient(cmap='Reds')
```

Observations: </br>
weak: values < 0.25 </br>
moderate: 0.25 < values < 0.75 </br>
strong: values > 0.75 </br>

1) Culmen Length and Flipper Length </br>
There is a moderate correlation between Culmen Length and Flipper Length (within all 3 species, all 3 values are close to 0.5).

2) Culmen Length and Body Mass </br>
There is a moderate coorelation between Culmen Length and Body Mass (within all 3 species,  all 3 values are close to 0.5)

3) Culment Depth and Delta 15 </br>
There is weak coorelation between  Culment Depth and Delta 15 (within all 3 species, all 3 values are less than 0.25)

4) Flipper Length and Body Mass </br>
There is a moderate correlation between Flipper Length and Body Mass (within all 3 species, all 3 values are close to 0.5), this correlation is the strongest among the 5 other relations

5) Delta 15 and Delta 13 </br>

# 5. K-mean clustering

```python
data_to_clfl = penguin_data.loc[:344, ["Culmen Length (mm)" ,"Flipper Length (mm)"]]
sns.pairplot(data_to_clfl);
```

```python
# scale the data using minmax scale
scaled_data = minmax_scale(data_to_clfl)

# initialize KMeans object with number of desired clusters
k_means = KMeans(n_clusters=3)

# fit and transform k_means object to scaled data
k_means.fit_transform(scaled_data)

# extract cluster labels
data_to_clfl["cluster_labels"] = k_means.labels_

# plot the clusters
sns.scatterplot(data=data_to_clfl, x="Culmen Length (mm)", y="Flipper Length (mm)", hue="cluster_labels")
plt.title("K-means graph for culmen lenngth and flipper length");
```
